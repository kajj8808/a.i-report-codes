```python
import tensorflow as tf
import matplotlib.pylab as plt
import numpy as np
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

w = tf.Variable(tf.random.normal([1], dtype=tf.float32))
b = tf.Variable(tf.random.normal([1], dtype=tf.float32))

def y(x):
    return x * w + b

x_data = np.arange(-2.0, 2.0, 0.1)
y_data = y(x_data)
plt.plot(x_data, y_data, color='#9b59b6', marker="o" , linestyle="-" , label='1st degree')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pylab as plt
import numpy as np
import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE"

a = tf.Variable(tf.random.normal([1], dtype=tf.float32))
b = tf.Variable(tf.random.normal([1], dtype=tf.float32)) 
c = tf.Variable(tf.random.normal([1], dtype=tf.float32))
d = tf.Variable(tf.random.normal([1], dtype=tf.float32))

def y(x):
    return a * x ** 3 + b * x ** 2 + c * x + d

x_data = np.arange(-2.0, 2.0, 0.1)
y_data = y(x_data)
plt.plot(x_data, y_data, color='#9b59b6', marker="o" , linestyle="-" , label='3st degree')
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
